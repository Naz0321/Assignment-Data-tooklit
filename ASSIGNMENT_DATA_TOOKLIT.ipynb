**Q2. How does broadcasting work in NumPy?**

Ans. Broadcasting is a powerful feature in NumPy that allows operations between arrays of different shapes without explicitly replicating data. It enables efficient vectorized operations by automatically expanding smaller arrays to match the shape of larger ones.

How It Works

When performing operations between arrays of different shapes, NumPy follows these broadcasting rules:

1. If the arrays have different dimensions, the smaller array is padded with ones on the left side until both arrays have the same number of dimensions.


2. The shapes are compared element-wise:

If the dimensions are equal, they remain unchanged.

If one of the dimensions is 1, it is stretched to match the other.

If the dimensions are different and neither is 1, broadcasting is not possible, and an error occurs.

**Q6. Differences Between NumPy Arrays and Python Lists?**


Ans. Both NumPy arrays and Python lists can store collections of elements, but they have significant differences in terms of performance, flexibility, and functionality.

1. Performance (Speed and Efficiency)

NumPy Arrays: Faster and more memory-efficient because they store elements in contiguous memory blocks and use optimized C-based operations.

Python Lists: Slower because they store pointers to objects, requiring extra memory and processing overhead.

2. Data Type Consistency

NumPy Arrays: Store elements of the same data type, ensuring better memory efficiency and performance.

Python Lists: Can store elements of different data types (e.g., integers, floats, strings), but this makes them less efficient.

3. Functionality and Operations

NumPy Arrays: Support vectorized operations (element-wise operations without explicit loops), making calculations more efficient.

Python Lists: Require explicit loops for mathematical operations, making them slower.


4. Memory Usage

NumPy Arrays: More memory-efficient because they store data in a contiguous memory block.

Python Lists: Less efficient due to storing references to objects.

5. Built-in Mathematical Functions

NumPy Arrays: Provide a wide range of mathematical, statistical, and scientific functions (np.mean(), np.sum(), np.std(), etc.).

Python Lists: Require manual implementation using loops or list comprehensions.

6. Multidimensional Support

NumPy Arrays: Support multi-dimensional data (e.g., 2D and 3D arrays).

Python Lists: Nested lists can mimic multi-dimensional structures, but they are not optimized.

**Q21. Why is NumPy’s array slicing faster than Python’s list slicing?**

Ans. NumPy’s array slicing is significantly faster than Python’s list slicing due to efficient memory management, optimized C-based operations, and lack of deep copying. Here’s why:

1. NumPy Uses Views Instead of Copies

When slicing a NumPy array, it returns a view (a reference to the original data), not a copy.

In contrast, Python lists create a new list object when sliced.

2. NumPy Uses Contiguous Memory Layout (Cache Optimization)

NumPy stores arrays in contiguous memory blocks (C-style memory layout).

Python lists store elements as pointers to objects, causing slower access due to cache misses.

3. NumPy Operations Are Implemented in C (Avoiding Python Overhead)

NumPy is built on highly optimized C and Fortran libraries, avoiding Python’s slow loops and overhead.

Python lists involve dynamic memory allocation and type checking, making them slower.

4. NumPy Supports Multi-Dimensional Efficient Slicing

Python lists require nested loops for multi-dimensional slicing.

NumPy supports slicing directly on multi-dimensional arrays.



NumPy’s slicing is faster than Python lists due to:
✔ Views Instead of Copies → Avoids unnecessary memory allocation.

✔ Contiguous Memory Storage → Optimized for CPU cache access.

✔ C-based Optimizations → Avoids Python’s slow dynamic typing.

✔ Efficient Multi-Dimensional Slicing → Works without nested loops.

```python
# Q4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap.

# ANS. The correlation matrix shows the linear relationship between numerical variables in a dataset. You can compute it using Pandas and visualize it using Seaborn’s heatmap.

# 1. Calculate the Correlation Matrix

#  You can use the .corr() function in Pandas to compute the correlation matrix.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    "A": [1, 2, 3, 4, 5],
    "B": [2, 3, 4, 5, 6],
    "C": [5, 4, 3, 2, 1],
    "D": [10, 20, 30, 40, 50]
}

df = pd.DataFrame(data)

# Compute Correlation Matrix
corr_matrix = df.corr()
print(corr_matrix)


#  2. Visualize the Correlation Matrix with a Heatmap

#  You can use Seaborn’s heatmap() to create a color-coded matrix.

# Create Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)

# Show Plot
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Q11. Perform matrix multiplication using NumPy.

# ANS. You can use NumPy to perform matrix multiplication using np.dot() or the @ operator.


import numpy as np

# Define two matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Matrix multiplication using np.dot()
result1 = np.dot(A, B)

# Matrix multiplication using @ operator
result2 = A @ B

# Print results
print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Result using np.dot():\n", result1)
print("Result using @ operator:\n", result2)
```
